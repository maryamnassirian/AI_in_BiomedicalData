## 4.3 Break, Continue
Break: We use break when we like to get out of or stop the loop.
Continue: With the continue statement we can skip the current iteration, and continue with the next:

```python
count = 0
while count < 5:
    print(count)
    count = count + 1
    if count == 3:
        break
```

```python
count = 0
while count < 5:
  print(count)
  if count == 3:
      count = count + 1
      continue
  count = count + 1
  print('Next number should be ', count)
print('outside the loop')
```

## 5.3 Inheritance

```python
class Person:
    def __init__(self, name, age, city):
        self.name = name
        self.age = age
        self.city = city

    # Method to update the city of residence
    def update_city(self, new_city):
        self.city = new_city
        return f"City updated to: {self.city}"

    # Method to check if the person is a senior (age >= 65)
    def is_eldery(self):
        return self.age >= 65

    def __getitem__(self, key):
        if key == 'info':
            return f"{self.name}, {self.age} years old. City: {self.city}."
        else:
            raise KeyError(f"Key '{key}' is not valid. Use 'info' to get person information.")

# Example usage of base class
person = Person("Mohammad Rezaei", 45, "Tehran")
print(person['info'])  # Output: John Doe, 45 years old. City: New York.

# Checking if the person is a senior
print(person.is_eldery())  # Output: False

# Updating the city
print(person.update_city("Rasht"))  # Output: City updated to: Los Angeles


```

```python
class Patient(Person):
    def __init__(self, name, age, city, medical_history=None):
        super().__init__(name, age, city)
        if medical_history is None:
            medical_history = []
        self.medical_history = medical_history

    # Method to add a new condition to medical history
    def add_medical_history(self, condition):
        self.medical_history.append(condition)
        return f"Added '{condition}' to medical history."

    # Method to get a summary of the patient's medical history
    def medical_summary(self):
        if self.medical_history:
            return f"Medical history: {self.medical_history}"
        else:
            return "No medical history recorded."

    def __getitem__(self, key):
        if key == 'info':
            base_info = super().__getitem__('info')
            return f"{base_info} Medical history: {self.medical_history}."
        else:
            return super().__getitem__(key)


patient = Patient("Reza Rezaei", 75, "Tehran", ["Cancer", "Diabetes"])
print(patient['info'])

print(patient.is_eldery())

print(patient.update_city("Rasht"))

print(patient.add_medical_history("Alzheimer"))

print(patient.medical_summary())
```
